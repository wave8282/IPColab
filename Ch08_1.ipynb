```python
from google.colab import drive
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
import matplotlib.pyplot as plt

drive.mount('/content/drive')
```

```python
img = cv2.imread('/content/drive/MyDrive/IP/Images/char.bmp')
print(img.shape)
gImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gImg, cmap='gray')
# cv2_imshow(gImg)
```

```python
kernel = np.ones((7,7), np.uint8)
dilateImg = cv2.dilate(gImg, kernel, iterations=1)
cv2_imshow(dilateImg)
```

```python
edgeImg = dilateImg - gImg
dilateImg = cv2.dilate(edgeImg, kernel, iterations=1)
cv2_imshow(dilateImg)
```

```python
img = cv2.imread('/content/drive/MyDrive/IP/CH08/M.bmp')
print(img.shape)
gImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gImg, cmap='gray')
# cv2_imshow(gImg)
```

```python
kernel = np.ones((3,3), np.uint8)
dilateImg = cv2.dilate(gImg, kernel, iterations=2)
cv2_imshow(dilateImg)
```

```python
edgeImg = dilateImg - gImg
dilateImg = cv2.dilate(edgeImg, kernel, iterations=1)
cv2_imshow(dilateImg)
```

```python
gBin = np.zeros((gImg.shape[0], gImg.shape[1]), dtype = np.uint8)
```

```python
gBin = np.zeros((gImg.shape[0], gImg.shape[1]), dtype = np.uint8)

threshold = 150

for i in range(gImg.shape[0]):
  for j in range(gImg.shape[1]):
    if gImg[i,j] > threshold:
      gBin[i,j] = 255

cv2_imshow(gBin)
```

```python
kernel = np.ones((3,3), np.uint8)
dilateImg = cv2.dilate(gBin, kernel, iterations=3)
cv2_imshow(dilateImg)
```

```python
#kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
#kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3,3))
kernel = cv2.getStructuringElement(cv2.MORPH_CROSS, (3,3))
dilateImg = cv2.dilate(gBin, kernel, iterations=3)
cv2_imshow(dilateImg)
```
