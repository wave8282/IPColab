```python
from google.colab import drive
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
import matplotlib.pyplot as plt

drive.mount('/content/drive')
```

```python
#img = cv2.imread('/content/drive/MyDrive/IP/CH05/L.jpg')
#img = cv2.imread('/content/drive/MyDrive/IP/CH05/rice.bmp')
img = cv2.imread('/content/drive/MyDrive/IP/CH05/p.jpg')
#img = cv2.imread('/content/drive/MyDrive/IP/CH05/M.bmp')
print(img.shape)
gImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# plt.imshow(gImg1, cmap='gray')
cv2_imshow(gImg)
```

```python
outImg =  cv2.blur(gImg, (7,7))
cv2_imshow(outImg)
```

```python
outImg =  cv2.GaussianBlur(gImg, (7,7), 5)
cv2_imshow(outImg)
```

```python
outImg =  cv2.medianBlur(gImg, 11)
cv2_imshow(outImg)
```

```python
outImage = gImg.copy()
# blurImage = cv2.GaussianBlur(outImage, (7,7), 3)
blurImage = cv2.medianBlur(outImage, 7)

x = 77
y = 75
size = 7

for h in range(y, y+size):
  for w in range(x, x+size):
    # outImage[h][w] = 255
    outImage[h][w] = blurImage[h][w]

cv2_imshow(outImage)
```

```python
#img = cv2.imread('/content/drive/MyDrive/IP/CH05/L.jpg')
#img = cv2.imread('/content/drive/MyDrive/IP/CH05/rice.bmp')
img = cv2.imread('/content/drive/MyDrive/IP/CH05/j.jpg')
#img = cv2.imread('/content/drive/MyDrive/IP/CH05/M.bmp')
print(img.shape)
gImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# plt.imshow(gImg1, cmap='gray')
cv2_imshow(gImg)
```

```python
outImg =  cv2.GaussianBlur(gImg, (7,7), 2)
cv2_imshow(outImg)
```

```python
edgeImage = cv2.addWeighted(gImg, 1, outImg, -1, 0)
cv2_imshow(edgeImage)
```

```python
unsharpImage = cv2.addWeighted(gImg, 1, edgeImage, 2, 0)
cv2_imshow(unsharpImage)
```

```python
kernel3 = np.ones((3,3), np.float32)/9
kernel3[0][0] = -1/9
kernel3[0][1] = -1/9
kernel3[0][2] = -1/9
kernel3[1][0] = -1/9
kernel3[1][1] =  17/9
kernel3[1][2] = -1/9
kernel3[2][0] = -1/9
kernel3[2][1] = -1/9
kernel3[2][2] = -1/9
```
