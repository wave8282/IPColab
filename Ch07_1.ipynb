```python
from google.colab import drive
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
import matplotlib.pyplot as plt

drive.mount('/content/drive')
```

```python
img = cv2.imread('/content/drive/MyDrive/IP/CH07/C.bmp')
print(img.shape)
gImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gImg, cmap='gray')
# cv2_imshow(gImg)
```

```python
outImg = np.zeros(gImg.shape, dtype=np.uint8)
#cv2_imshow(outImg)
```

```python
scaling = 1.5

for h in range(gImg.shape[0]):
  for w in range(gImg.shape[1]):
    y = int(h*scaling)
    x = int(w*scaling)
    if y < gImg.shape[0] and x < gImg.shape[1]:
        outImg[y, x] = gImg[h, w]

cv2_imshow(outImg)
```

```python
scaling = 3

for y in range(gImg.shape[0]):
  for x in range(gImg.shape[1]):
    h = int(y/scaling)
    w = int(x/scaling)
    if h < gImg.shape[0] and w < gImg.shape[1]:
        outImg[y, x] = gImg[h, w]

cv2_imshow(outImg)
```

```python
#outImg = cv2.resize(gImg, None, fx=3, fy=3, interpolation=cv2.INTER_NEAREST)
#outImg = cv2.resize(gImg, None, fx=3, fy=3, interpolation=cv2.INTER_LINEAR)
outImg = cv2.resize(gImg, None, fx=3, fy=3, interpolation=cv2.INTER_CUBIC)
cv2_imshow(outImg)
```

```python
height, width = gImg.shape
print(height, width)
```

```python
M = cv2.getRotationMatrix2D((width/2.0, height/2.0), 45, 1)
print(M)
```

```python
outImg = cv2.warpAffine(gImg, M, (width, height))
cv2_imshow(outImg)
```
