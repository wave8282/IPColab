```python
from google.colab import drive
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
import matplotlib.pyplot as plt

drive.mount('/content/gdrive')
```

```python
img = cv2.imread('/content/gdrive/MyDrive/IP/CH03/pep.jpg')
print(img.shape)
gImg1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gImg1, cmap='gray')
```

```python
def saturation(val):
  if val > 255:
    return 255
  elif val < 0:
    return 0
  else:
    return val
```

```python
gPoint = np.zeros((gImg1.shape[0], gImg1.shape[1]), dtype=np.ubyte)
```

```python
for h in range(gImg1.shape[0]):
  for w in range(gImg1.shape[1]):
    pointVal = 255 - np.int32(gImg1[h, w])
    gPoint[h, w] = saturation(pointVal)

plt.imshow(gPoint, cmap='gray')
```

이진화


```python
th = 30


for h in range(gImg1.shape[0]):
  for w in range(gImg1.shape[1]):
   if gImg1[h, w] > th:
     gPoint[h, w] = 255
   else:
     gPoint[h, w] = 0
plt.imshow(gPoint, cmap='gray')
```

```python
numOfClip = 4
modVal = 256/numOfClip


for h in range(gImg1.shape[0]):
  for w in range(gImg1.shape[1]):
    clipVal = np.int32(gImg1[h, w])% modVal
    gPoint[h, w] = clipVal



plt.imshow(gPoint, cmap='gray')
```
