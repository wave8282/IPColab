```python
from google.colab import drive
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
import matplotlib.pyplot as plt

drive.mount('/content/drive')
```

```python
img = cv2.imread('/content/drive/MyDrive/IP/CH08/p.bmp')
print(img.shape)
gImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gImg, cmap='gray')
# cv2_imshow(gImg)
```

```python
kernel = np.ones((3, 3), np.uint8)
erosion = cv2.erode(gImg, kernel, iterations=1)
cv2_imshow(erosion)
```

```python
kernel = np.ones((3, 3), np.uint8)
opening = cv2.dilate(erosion, kernel, iterations=1)
cv2_imshow(opening)
```

```python
dilate2 = cv2.dilate(opening, kernel, iterations=1)
cv2_imshow(dilate2)
```

```python
erode2 = cv2.erode(dilate2, kernel, iterations=1)
cv2_imshow(erode2)
```
