```python
#img = cv2.imread('/content/drive/MyDrive/IP/CH05/L.jpg')
#img = cv2.imread('/content/drive/MyDrive/IP/CH05/rice.bmp')
img = cv2.imread('/content/drive/MyDrive/IP/CH05/j.jpg')
#img = cv2.imread('/content/drive/MyDrive/IP/CH05/M.bmp')
print(img.shape)
gImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# plt.imshow(gImg1, cmap='gray')
cv2_imshow(gImg)
```

```python
def saturation(value):
  if(value > 255):
    value = 255
  elif(value < 0):
    value = 0
  return value
```

```python
kernel3 = np.ones((3,3), np.float32)/9
```

```python
output = cv2.filter2D(gImg, -1, kernel3)
cv2_imshow(output)
```

```python
output = cv2.filter2D(output, -1, kernel3)
cv2_imshow(output)
```

```python
gOut = np.zeros((gImg.shape[0], gImg.shape[1]), dtype=np.ubyte)
```

```python
for h in range(1, gImg.shape[0]-1):
  for w in range(1, gImg.shape[1]-1):
    sum = 0
    for i in range(-1, 2):
      for j in range(-1, 2):
        sum += gImg[h+i, w+j] * kernel3[i+1, j+1]
    sum = saturation(sum)
    gOut[h, w] = sum

cv2_imshow(gOut)
```
